```python
import pandas as pd
import geopandas as gpd
import json
from bokeh.models import GeoJSONDataSource, LinearColorMapper
from bokeh.plotting import figure
from bokeh.io import show
from bokeh.palettes import brewer
```

```python
df = pd.read_csv("/home/scharf/Desktop/DS_Data_Visualisation_2022_Vanessa_Scharf/Lyme/data/aggregated_states.csv")
df
```

```python
df.drop('Stname.1', axis=1, inplace=True)
```

```python
shapefile = '/home/scharf/Desktop/DS_Data_Visualisation_2022_Vanessa_Scharf/Lyme/natural_earth_vector/10m_cultural/ne_10m_admin_1_states_provinces.shp'
gdf = gpd.read_file(shapefile)[['name', 'geometry']]
gdf.columns = ['State', 'geometry']
gdf
```

```python
merged_df = gdf.merge(df, right_on = 'Stname', left_on = 'State')
merged_df
```

```python
merged_json = json.loads(merged_df.to_json())
type(merged_json)
```

```python
json_data = json.dumps(merged_json)
type(json_data)
```

```python
geo_source = GeoJSONDataSource(geojson = json_data)
palette = brewer['OrRd'][8]
color_mapper = LinearColorMapper(palette=palette)
ghg_map = figure(title='Lyme Disease',plot_height=1200, plot_width=2200, )
ghg_map.patches(source=geo_source, fill_color={'field':'Count', 'transform':color_mapper}
               )
```

```python
show(ghg_map)
```
