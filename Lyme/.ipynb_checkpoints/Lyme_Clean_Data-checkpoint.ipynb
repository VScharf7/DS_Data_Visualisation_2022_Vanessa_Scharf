```python
import pandas as pd
import numpy as np
import plotly.express as px  # (version 4.7.0 or higher)
import plotly.graph_objects as go
from dash import Dash, dcc, html, Input, Output  # pip install dash (version 2.0.0 or higher)
```

```python
lyme = pd.read_csv('LD-Case-Counts-by-County-00-19_copy.csv')
lyme.head()
```

```python
lyme.drop('Ctyname', axis=1, inplace=True)
#lyme.replace(0, np.nan, inplace=True)
lyme.head()
```

```python
# States aggregieren
```

```python
aggregated_states = {'2000': 'sum', '2001': 'sum', '2002': 'sum', '2003': 'sum', '2004': 'sum',
                     '2005': 'sum', '2006': 'sum', '2007': 'sum', '2008': 'sum', '2009': 'sum',
                     '2010': 'sum', '2011': 'sum', '2012': 'sum', '2013': 'sum', '2014': 'sum',
                     '2015': 'sum', '2016': 'sum', '2017': 'sum', '2018': 'sum', '2019': 'sum', 'Stname': 'first'}
```

```python
df_new = lyme.groupby(lyme['Stname']).aggregate(aggregated_states)
df_new.head()
```

```python
df_new.to_csv('aggregated_states.csv')
```
