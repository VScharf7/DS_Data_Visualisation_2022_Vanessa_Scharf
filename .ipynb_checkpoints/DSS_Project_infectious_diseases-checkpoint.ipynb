```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ID = pd.read_csv("/home/scharf/Desktop/DS_Data_Visualisation_2022_Vanessa_Scharf/infectious-diseases-by-county-year-and-sex.csv")
ID.head()
```

```python
#list all diseases
ID.Disease.unique()
```

```python
ID.groupby(['Disease', 'Year', 'County']).mean()
```

```python
Lyme = ID[ID.Disease == "Lyme Disease"]
Lyme.head()
```

```python
lyme_count = Lyme[Lyme.Cases > 0]
lyme_count.head()
```

```python
sns.displot(data=lyme_count, y= "Cases", x= "County", hue="Year")
```

```python
sns.lineplot(data=lyme_count, y="Cases", x="Year", hue="Sex")
```
