```python
import pandas as pd
import random
```

```python
data = pd.read_csv("infectious-diseases-by-county-year-and-sex.csv")
```

```python
data
```

```python
data.groupby(['Disease', 'Year', 'County']).mean()
```

```python
lyme = data[data.Disease == 'Lyme Disease']
```

```python
lyme_greater_zero = lyme[lyme.Cases > 0]
lyme_greater_zero.head()
```

```python
lyme_greater_zero.columns
```

```python
lyme_cleaned = lyme_greater_zero.drop(lyme_greater_zero.columns[[6,7,8,]], axis=1)
new= lyme_cleaned.replace('TOTAL', '')
```

```python
import panel as pn
from bokeh.models import ColumnDataSource
from bokeh.io import curdoc
from bokeh.layouts import row
import pandas_bokeh
```

```python
# Festlegen des Dashboard Titels
curdoc().title = "Lyme Disease Cases"
# Datenset laden
```

```python
new.groupby(['Year', 'County']).mean()
```

```python
# Plot1 - Line plot
p_line= new.groupby(['Cases', 'County']).mean().plot_bokeh(kind="line",y="Year",color='#d01c8b',plot_data_points=True,show_figure=True)
# Plot2- Barplot
p_bar = new.groupby(['Cases', 'County']).mean().plot_bokeh(kind="bar",show_figure=False)
```

```python
# Plot3 - Scatterplot
p_scatter = new.plot_bokeh(kind="scatter", x="County", y="Cases",category="Year",show_figure=True)
```

```python
# Make Dashboard with Grid Layout
pandas_bokeh.plot_grid([p_line], width=400)
```
